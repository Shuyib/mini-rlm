# mini-rlm

Notebook-first implementation of recursive language model utilities exported with nbdev.

## Main Notebook

- [rlm_lisette_v2](rlm_lisette_v2.ipynb)

## Quickstart

Install the package:

```bash
pip install mini-rlm
```

Set your provider key before running recursive LLM calls:

```bash
export OPENROUTER_API_KEY=your_key_here
```

Import the package and call `rlm` with a query and a context string:

```python
from mini_rlm.rlm_lisette_v2 import rlm

context = """
Ada Lovelace is often regarded as the first computer programmer.
Grace Hopper popularized machine-independent programming languages.
"""

answer = rlm(
    query="Who is described as the first computer programmer?",
    context=context,
    root_model="openai/gpt-oss-120b",
    sub_model="openai/gpt-oss-20b",
)

print(answer)
```

If you want to work with the lower-level class directly:

```python
from mini_rlm.rlm_lisette_v2 import RLM

context = "Paris is the capital of France."
agent = RLM(context=context)
result = agent("What is the capital of France?", max_steps=2)

print(result)
```

If you only want to prepare the REPL shell and inspect the injected context/tooling:

```python
from mini_rlm.rlm_lisette_v2 import prep_shell

shell = prep_shell("Example context for inspection")
print(shell.user_ns["context"])
print(callable(shell.user_ns["llm_query"]))
```
